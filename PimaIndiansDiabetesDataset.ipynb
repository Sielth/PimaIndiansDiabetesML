```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Retrieves the dataset and gives every column a name.

df = pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv')
df.columns = ['Number of times pregnant','Plasma glucose concentration','Diastolic blood pressure','Triceps skinfold thickness','2-Hour serum insulin','Body mass index','Diabetes pedigree function','Age','Diabetes (0/1)']
```

```python
# Prints the dataframe

print(df)
```

```python
# Creates a correlation matrix with Pandas.

corrMatrix = df.corr()
print(corrMatrix)
```

```python
# Creates a visual representation of the correlation matrix using Seaborn's heatmap.

sns.heatmap(corrMatrix)
plt.show()
```

```python
# Creates a visual representation of the correlation matrix using Seaborn's heatmap.

# Return a copy of a matrix with the elements below the k-th diagonal zeroed.
mask = np.triu(np.ones_like(corrMatrix, dtype=bool))

# Create a figure and a set of subplots.
f, ax = plt.subplots(figsize=(11, 9))

# Make a diverging palette between two HUSL colors.
cmap = sns.diverging_palette(1, 150, s=85, n=9, center="dark", as_cmap=True)

sns.heatmap(corrMatrix, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

```python
# Creates a Panda's scatter matrix plot.

axes = pd.plotting.scatter_matrix(df, alpha=0.2)

# Aligns the y label text.
for ax in axes.flatten():
    ax.xaxis.label.set_rotation(90)
    ax.yaxis.label.set_rotation(0)
    ax.yaxis.label.set_ha('right')

plt.show()
```
